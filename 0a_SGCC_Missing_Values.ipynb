```
# Drop the row if all of the values in the row are missing.
dataset.dropna(how='all',inplace=True)

# Missing values percentage
round((dataset.isnull().sum().sum() / (dataset.shape[0] * (dataset.shape[1]-2))) * 100.00, 2)
```

```
# Drop missing values if more than 50% by row (axis=0)
dataset.dropna(thresh=dataset.columns.size/2, axis=0, inplace=True)

# Missing values along the row after dropped
dataset.isnull().sum(axis=1)
```

```
# Missing values along the columns after dropped
dataset.isnull().sum(axis=0)
```

```
# Missing values percentage
round((dataset.isnull().sum().sum() / (dataset.shape[0] * (dataset.shape[1]-2))) * 100.00, 2)
```

```
# Drop missing values if more than 50% by columns (axis=1)
dataset.dropna(thresh=len(dataset.index)/2, axis=1, inplace=True)

# Missing values along the columns
dataset.isnull().sum(axis=0)
```

```
# Missing values percentage
round((dataset.isnull().sum().sum() / (dataset.shape[0] * (dataset.shape[1]-2))) * 100.00, 2)
```

```
# Convert all date columns into correct datetime format and sort
dataset_new = dataset.copy()
dataset_new = dataset.iloc[:,2:]
dataset_new.columns = pd.to_datetime(dataset_new.columns).date
dataset_new = dataset_new.reindex(sorted(dataset_new.columns), axis=1)
dataset_new['FLAG'] = dataset['FLAG']
dataset_new['CONS_NO'] = dataset['CONS_NO']
dataset_new = dataset_new.set_index('FLAG').reset_index()
dataset_new
```

```
# Select date columns only
dataset_new_date = dataset_new.iloc[:, 1:-1]
dataset_new_date
```

```
# Replace missing values with previous or after consumption day value by row
dataset_new_date.fillna(method='bfill', axis=1, inplace=True)
dataset_new_date.fillna(method='ffill', axis=1, inplace=True)
dataset_new_date
```

```
# No more missing values
dataset_new_date.isnull().sum().sum()
```
